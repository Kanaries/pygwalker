```python
data = [{"season": "winter", "casual_sum": 48698}, {"season": "spring", "casual_sum": 180531}, {"season": "fall", "casual_sum": 166686}, {"season": "summer", "casual_sum": 224102}]

import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame(data)
df.plot(x='season', y='casual_sum', kind="bar")
plt.xlabel('season')
plt.ylabel('sum(casual)')
plt.show()
```
